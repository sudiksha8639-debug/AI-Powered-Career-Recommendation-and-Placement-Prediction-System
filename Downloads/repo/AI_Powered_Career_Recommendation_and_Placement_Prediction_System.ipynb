# 🛠️ Data Preprocessing

Data preprocessing is one of the most important phases in any Machine Learning project.

Raw datasets often contain unnecessary features, categorical variables, and missing values that must be handled before training a machine learning model.

In this section, the dataset is cleaned and transformed into a format suitable for machine learning algorithms.

# 🗑️ Removing Unnecessary Columns

The **sl_no** column represents only the serial number of each student.

Since it does not contribute to predicting placement status, it is removed from the dataset.

```
df = df.drop(columns=['sl_no'])

df.head()
```

## 📌 Observation

The serial number column has been successfully removed because it does not contain predictive information.

Removing irrelevant features helps improve model efficiency and prevents unnecessary complexity.

# 💰 Handling Missing Values

The **salary** column contains missing values for students who were not placed.

Since salary is unavailable before placement and would introduce data leakage, it is excluded from the prediction model.

```
df = df.drop(columns=['salary'])

df.head()
```

## 📌 Observation

The salary column has been removed because it is not available at the time of prediction and is directly dependent on placement status.

Excluding this feature prevents data leakage and improves the fairness of the predictive model.

# 🔤 Encoding Categorical Variables

Machine Learning algorithms cannot directly process categorical (text) values.

Therefore, all categorical features are converted into numerical form using Label Encoding.

```
label_encoder = LabelEncoder()

categorical_columns = df.select_dtypes(include='object').columns

categorical_columns
```

```
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

df.head()
```

## 📌 Observation

All categorical features have been successfully transformed into numerical values using Label Encoding.

The dataset is now compatible with machine learning algorithms that require numerical input.

# 📊 Encoded Dataset Information

# 🎯 Defining Features and Target Variable

Machine Learning requires separating the independent variables (features) from the dependent variable (target).

In this project:

- **X** represents all input features.
- **y** represents the placement status.

```
X = df.drop('status', axis=1)

y = df['status']

print("Features Shape :", X.shape)
print("Target Shape :", y.shape)
```

## 📌 Observation

The independent features and target variable have been successfully separated.

The model will use the feature matrix **X** to predict the target variable **y**.

# ✂️ Splitting the Dataset

To evaluate model performance fairly, the dataset is divided into:

- Training Set (80%)
- Testing Set (20%)

The training set is used to train the model, while the testing set is used to evaluate its predictive performance.

```
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Features :", X_train.shape)
print("Testing Features :", X_test.shape)

print()

print("Training Labels :", y_train.shape)
print("Testing Labels :", y_test.shape)
```

## 📌 Observation

The dataset has been successfully divided into training and testing subsets.

Using an 80:20 split ensures that sufficient data is available for training while maintaining an independent test set for evaluating model performance.

# 📈 Feature Scaling

Some machine learning algorithms, such as Logistic Regression and Support Vector Machine (SVM), perform better when numerical features are standardized.

Feature scaling ensures that all numerical features contribute equally during model training.

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)

print("Feature scaling completed successfully.")
```

## 📌 Observation

The numerical features have been standardized using StandardScaler.

Feature scaling improves the convergence and performance of distance-based and optimization-based machine learning algorithms.

# 🤖 Machine Learning Model Development

In this section, multiple supervised machine learning algorithms are trained and evaluated for predicting student placement status.

Instead of relying on a single algorithm, multiple classification models are compared to determine the most suitable model for the given dataset.

The following algorithms are implemented:

- Logistic Regression
- Decision Tree Classifier
- Random Forest Classifier
- Support Vector Machine (SVM)

Each model is evaluated using standard classification metrics to ensure a fair comparison.

# 📌 Model 1: Logistic Regression

Logistic Regression is one of the most widely used supervised learning algorithms for binary classification problems.

It predicts the probability of a student being placed or not placed based on the input features.

Since Logistic Regression is sensitive to feature scales, the standardized dataset is used.

```
log_model = LogisticRegression(random_state=42)

log_model.fit(X_train_scaled, y_train)

log_pred = log_model.predict(X_test_scaled)

log_accuracy = accuracy_score(y_test, log_pred)

print(f"Logistic Regression Accuracy: {log_accuracy:.2%}")
```

```
print(classification_report(y_test, log_pred))
```

```
cm = confusion_matrix(y_test, log_pred)

plt.figure(figsize=(6,5))

sns.heatmap(cm,
            annot=True,
            fmt='d',
            cmap='Blues')

plt.title("Logistic Regression Confusion Matrix")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()
```

## 📌 Observation

Logistic Regression establishes a strong baseline model for placement prediction.

The confusion matrix indicates how effectively the model distinguishes between placed and non-placed students.

# 🌳 Model 2: Decision Tree Classifier

Decision Trees classify data by learning a sequence of decision rules.

Unlike Logistic Regression, Decision Trees do not require feature scaling and can naturally capture non-linear relationships.

```
dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train, y_train)

dt_pred = dt_model.predict(X_test)

dt_accuracy = accuracy_score(y_test, dt_pred)

print(f"Decision Tree Accuracy: {dt_accuracy:.2%}")
```

```
print(classification_report(y_test, dt_pred))
```

```
cm = confusion_matrix(y_test, dt_pred)

plt.figure(figsize=(6,5))

sns.heatmap(cm,
            annot=True,
            fmt='d',
            cmap='Greens')

plt.title("Decision Tree Confusion Matrix")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()
```

## 📌 Observation

The Decision Tree model captures non-linear decision boundaries effectively.

However, individual decision trees may overfit the training data if not properly controlled.

# 🌲 Model 3: Random Forest Classifier

Random Forest is an ensemble learning algorithm that combines multiple decision trees.

By aggregating predictions from many trees, Random Forest generally provides higher accuracy and better generalization than a single Decision Tree.

```
rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_test)

rf_accuracy = accuracy_score(y_test, rf_pred)

print(f"Random Forest Accuracy: {rf_accuracy:.2%}")
```

```
print(classification_report(y_test, rf_pred))
```

```
cm = confusion_matrix(y_test, rf_pred)

plt.figure(figsize=(6,5))

sns.heatmap(cm,
            annot=True,
            fmt='d',
            cmap='Oranges')

plt.title("Random Forest Confusion Matrix")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()
```

## 📌 Observation

Random Forest reduces overfitting by combining multiple decision trees.

It often achieves one of the highest accuracies for structured tabular datasets such as placement prediction.

# ⚡ Model 4: Support Vector Machine (SVM)

Support Vector Machine is a powerful supervised learning algorithm that finds the optimal boundary separating different classes.

Since SVM is distance-based, standardized features are used for training.

```
svm_model = SVC(random_state=42)

svm_model.fit(X_train_scaled, y_train)

svm_pred = svm_model.predict(X_test_scaled)

svm_accuracy = accuracy_score(y_test, svm_pred)

print(f"SVM Accuracy: {svm_accuracy:.2%}")
```

```
print(classification_report(y_test, svm_pred))
```

```
cm = confusion_matrix(y_test, svm_pred)

plt.figure(figsize=(6,5))

sns.heatmap(cm,
            annot=True,
            fmt='d',
            cmap='Purples')

plt.title("SVM Confusion Matrix")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()
```

## 📌 Observation

Support Vector Machine performs well for classification tasks involving clear decision boundaries.

Feature scaling significantly improves its predictive performance.

# 🏆 Model Performance Comparison

After training multiple Machine Learning models, it is essential to compare their performance using various evaluation metrics.

The models are evaluated based on:

- Accuracy
- Precision
- Recall
- F1-Score

Comparing multiple algorithms helps identify the most suitable model for placement prediction.

```
results = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Decision Tree",
        "Random Forest",
        "Support Vector Machine"
    ],
    "Accuracy": [
        accuracy_score(y_test, log_pred),
        accuracy_score(y_test, dt_pred),
        accuracy_score(y_test, rf_pred),
        accuracy_score(y_test, svm_pred)
    ],
    "Precision": [
        precision_score(y_test, log_pred),
        precision_score(y_test, dt_pred),
        precision_score(y_test, rf_pred),
        precision_score(y_test, svm_pred)
    ],
    "Recall": [
        recall_score(y_test, log_pred),
        recall_score(y_test, dt_pred),
        recall_score(y_test, rf_pred),
        recall_score(y_test, svm_pred)
    ],
    "F1 Score": [
        f1_score(y_test, log_pred),
        f1_score(y_test, dt_pred),
        f1_score(y_test, rf_pred),
        f1_score(y_test, svm_pred)
    ]
})

results
```

## 📌 Observation

The above table summarizes the performance of all machine learning models.

A comparative evaluation provides better insight than relying solely on accuracy and helps identify the most balanced model.

# 📊 Accuracy Comparison

A visual comparison of model accuracies helps quickly identify the best-performing algorithm.

```
plt.figure(figsize=(9,6))

bars = plt.bar(results["Model"], results["Accuracy"])

plt.title("Comparison of Machine Learning Models")

plt.xlabel("Models")

plt.ylabel("Accuracy")

plt.ylim(0.5,1)

for bar in bars:
    plt.text(
        bar.get_x()+bar.get_width()/2,
        bar.get_height()+0.01,
        f"{bar.get_height():.2f}",
        ha='center'
    )

plt.show()
```

## 📌 Observation

The bar chart clearly illustrates the comparative performance of all classification models.

The model with the highest accuracy demonstrates superior predictive capability on the testing dataset.

# 🔄 K-Fold Cross Validation

A single train-test split may produce optimistic or pessimistic performance depending on how the data is divided.

To obtain a more reliable estimate of model performance, K-Fold Cross Validation is performed.

In this project, 5-fold cross validation is used, where the dataset is divided into five equal subsets. The model is trained on four subsets and tested on the remaining subset. This process is repeated five times, and the average accuracy is reported.

```
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(
    rf_model,
    X,
    y,
    cv=5,
    scoring='accuracy'
)

print("Cross Validation Scores")

print(cv_scores)

print()

print("Average Accuracy :", cv_scores.mean())

print("Standard Deviation :", cv_scores.std())
```

## 📌 Observation

The cross-validation scores indicate that the Random Forest model performs consistently across different subsets of the dataset.

A low standard deviation suggests that the model generalizes well and is not heavily dependent on a particular train-test split.

# 🎯 Hyperparameter Tuning using GridSearchCV

Machine learning models contain several parameters that influence their performance.

Instead of selecting these parameters manually, GridSearchCV systematically searches different parameter combinations to identify the optimal configuration.

```
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

best_rf_model = grid_search.best_estimator_

print("Best Parameters")

print(grid_search.best_params_)

print()

print("Best Cross Validation Accuracy")

print(grid_search.best_score_)
```

## 📌 Observation

GridSearchCV identified the combination of hyperparameters that achieved the highest cross-validation accuracy.

The optimized Random Forest model is expected to provide better predictive performance than the default configuration.

# 🔍 Explainable Artificial Intelligence (XAI) using SHAP

Machine Learning models often achieve excellent predictive performance; however, they are frequently considered "black-box" models because their decision-making process is difficult to interpret.

To improve the transparency and interpretability of the proposed placement prediction system, SHAP (SHapley Additive exPlanations) has been incorporated. SHAP explains how each feature contributes to the final prediction by assigning an importance value to every feature.

This analysis helps identify the most influential factors affecting student placement and enhances the trustworthiness of the developed machine learning model.

```
!pip install shap
```

```
print(best_rf_model)
```

```
import shap
import numpy as np

# Create SHAP Explainer
explainer = shap.TreeExplainer(best_rf_model)

# Generate SHAP values
shap_values = explainer.shap_values(X_test)

# Handle different SHAP versions
if isinstance(shap_values, list):
    shap.summary_plot(shap_values[1], X_test)
elif isinstance(shap_values, np.ndarray) and shap_values.ndim == 3:
    shap.summary_plot(shap_values[:, :, 1], X_test)
else:
    shap.summary_plot(shap_values, X_test)
```

## 📈 SHAP Feature Importance Analysis

The SHAP Summary Plot provides a global interpretation of the trained Random Forest model by ranking all input features according to their contribution to placement prediction.

Unlike traditional feature importance methods, SHAP not only identifies the most influential features but also indicates whether each feature positively or negatively impacts the prediction. This makes the machine learning model more transparent, interpretable, and trustworthy.

```
import shap
import matplotlib.pyplot as plt
import numpy as np

# Handle different SHAP output formats
if isinstance(shap_values, list):
    values = shap_values[1] if len(shap_values) > 1 else shap_values[0]
elif isinstance(shap_values, np.ndarray) and shap_values.ndim == 3:
    values = shap_values[:, :, 1]
else:
    values = shap_values

shap.summary_plot(values, X_test, plot_type="bar")
```

## 📌 Observation

From the SHAP Feature Importance analysis, it is evident that only a few features contribute significantly towards predicting student placement. Features appearing at the top of the visualization have the highest influence on the model's decisions, whereas features near the bottom have comparatively lower impact.

The SHAP analysis provides valuable insights into the decision-making process of the Random Forest classifier and helps validate that the model is making predictions based on meaningful academic and employability-related attributes rather than random patterns.
